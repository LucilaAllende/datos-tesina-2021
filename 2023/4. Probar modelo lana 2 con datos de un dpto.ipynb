```python
from statsmodels.iolib.smpickle import load_pickle
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
modelo_lana = load_pickle('modelos/modelo_prediccion_lana_2')
```

## Prueba con una tupla

Se tomo las primeras dos filas del dataset *datos_dptp_ha.csv*, que son filas corresponfientes a Biedma.

| total_ovinos | kilos_lana | cant_lluvia |
| --- | --- | --- |
| 193849 | 649130.0 | 46.40 |
| 187897 | 620176.0 | 226.50 |

```python
resultado_prediccion = modelo_lana.predict([[46.40,193849], [226.50,187897]])
resultado_prediccion
```

```python
dato_real =[[649130.0],[620176.0]]
```

```python
print('Exactitud de la predicción:')
print(str(round(r2_score(dato_real, resultado_prediccion),4)))
```

```python
print('MSE:')
print(str(round(mean_squared_error(dato_real, resultado_prediccion),4)))
```

```python
plt.plot(range(len(dato_real)),dato_real,label="Real")
plt.plot(range(len(resultado_prediccion)),resultado_prediccion,label="Prediccion")
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
plt.xlabel('muestras')
plt.ylabel('kilos lana')
plt.legend()
plt.show()
```
